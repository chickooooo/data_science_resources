```python
# error line chart
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
lower_bound = [i - 0.1 * i for i in revenue]
upper_bound = [i + 0.1 * i for i in revenue]

plt.figure(figsize=(9, 5))
ax = plt.subplot()
plt.fill_between(months, lower_bound, upper_bound, color="red", alpha=0.1)
plt.plot(months, revenue, color="red")
ax.set_xticks(months)
ax.set_xticklabels(month_names)
plt.show()
```

bar chart

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales =  [91, 76, 56, 66, 52, 27]
ax = plt.subplot()
plt.bar(range(len(sales)), sales, color="pink")
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks, rotation=0)
plt.xlabel("drinks")
plt.ylabel("sales")
plt.title("Sales of drinks")
plt.show()
```

```python
# side by side bars
plt.figure(figsize=(9, 5))
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

n = 1  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
x_values1 = [t*element + w*n for element
             in range(d)]
n = 2
x_values2 = [t*element + w*n for element
             in range(d)]
x_ticks = [(i+j)/2 for i, j in zip(x_values1, x_values2)]

ax = plt.subplot()
plt.bar(x_values1, sales1, color="pink", label = "sales1")
plt.bar(x_values2, sales2, color="gray", label="sales2")
ax.set_xticks(x_ticks)
ax.set_xticklabels(drinks, rotation=0)
plt.xlabel("drinks")
plt.ylabel("sales")
plt.title("Sales of drinks")
plt.legend()

plt.show()
```

```python
# stacked bars
plt.figure(figsize=(9, 6))
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

ax = plt.subplot()
plt.bar(range(len(drinks)), sales1, color="orange", label = "sales1")
plt.bar(range(len(drinks)), sales2, bottom=sales1, color="pink", label = "sales2")
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks, rotation=0)
plt.xlabel("drinks")
plt.ylabel("sales")
plt.title("Sales of drinks")
plt.legend()

plt.show()
```

```python
# error bars
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]
error = [0.6, 0.9, 0.4, 0, 0.9, 0]

ax = plt.subplot()
plt.bar(range(len(drinks)), ounces_of_milk, yerr=error, capsize=10, color="orange", label="sales1")
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks)
plt.xlabel("drinks")
plt.ylabel("ounces of milk")
plt.title("Milk in drinks")
plt.legend()

plt.show()
```

```python
country_data = pd.read_csv("./country_data.tsv", delimiter="\t")
country_data.head()
```

Univariate quantitative variable

```python
# boxplot
sns.boxplot(data=country_data, x="lifeExp")
plt.show()
```

```python
# histogram
sns.histplot(data = country_data, x="lifeExp", bins=30, kde=False)
plt.show()
```

Univariate categorical variable

```python
# countplot / bar chart
sns.countplot(data=country_data, x="continent")
plt.show()
```

```python
# pie chart
plt.pie(x=country_data.continent.value_counts(), labels=country_data.continent.value_counts().index, autopct="%0.1f%%")
plt.show()
```

Bivariate variables: one quantitative, one categorical

```python
# side by side box plots
sns.boxplot(x="continent", y="lifeExp", data=country_data)
plt.show()
```

```python
# overlapping histograms
plt.figure(figsize=(8, 5))
plt.hist(x = country_data[country_data.continent == "Asia"].lifeExp, label="Asia", density=True, alpha=0.5)
plt.hist(x = country_data[country_data.continent == "Europe"].lifeExp, label="Europe", density=True, alpha=0.5)
plt.hist(x = country_data[country_data.continent == "Africa"].lifeExp, label="Africa", density=True, alpha=0.5)
plt.hist(x = country_data[country_data.continent == "Americas"].lifeExp, label="Americas", density=True, alpha=0.5)
plt.hist(x = country_data[country_data.continent == "Oceania"].lifeExp, label="Oceania", density=True, alpha=0.5)
plt.legend()
plt.show()
```

Bivariate variables: two quantitative

```python
# scatter plot
sns.scatterplot(data=country_data, x="lifeExp", y="gdpPercap")
plt.show()
```
